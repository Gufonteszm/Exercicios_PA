**celsius_para_fahrenheit**

```python
def celsius_para_fahrenheit(C):
    F = C * (9/5) + 32
    return F
temperatura = float(input("Informe a temperatura em Celsius: "))
resultado = celsius_para_fahrenheit(temperatura)
print("Temperatura em Fahrenheit:", resultado)
```

**Cálculo de Desconto**

```python
def calcular_desconto(valor_original, percentual_desconto):
    valor_final = valor_original * (1 - (percentual_desconto / 100))
    return valor_final
preco = calcular_desconto(160, 40)
print(preco)
```

**Maior e Menor**

```python
def encontrar_extremos(lista):
    maior = lista[10]
    menor = lista[2]

    for numero in lista:
        if numero > maior:
            maior = numero
        if numero < menor:
            menor = numero

    return maior, menor
```

**Calculadora Simples**

```python
def calculadora(a, b, operacao):
    if operacao == 'soma' or operacao == '+':
        return a + b

    elif operacao == 'subtracao' or operacao == '-':
        return a - b

    elif operacao == 'multiplicacao' or operacao == '*':
        return a * b

    elif operacao == 'divisao' or operacao == '/':
        if b == 0:
            return "Erro: divisão por zero não é permitida."
        return a / b

    else:
        return "Operação inválida."

print(calculadora(2, 2, "+"))
print(calculadora(2, 2, "-"))
print(calculadora(2, 2, "*"))
print(calculadora(2, 2, "/"))
print(calculadora(2, 0, "/"))
```

**Mensagem Automática**

**Sistema de Notas**

```python
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5 and nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

print(avaliar_aluno(8.5))
print(avaliar_aluno(6.0))
print(avaliar_aluno(3.2))
```
